```python
import numpy as np
from pathlib import Path
```

```python
# src_dir = Path("jobs/FLIP_human/")
# src_dir = Path("jobs/FLIP_mixed/")
src_dir = Path("jobs/ncRNA_dataset2/")
dataset_name = src_dir.stem
src_files = list(src_dir.glob("output/*.npy"))
src_files = sorted(src_files, key=lambda p: (len(str(p)), str(p)))
src_files
```

```python
blocks = [np.load(file) for file in src_files]
# lcs_matrix_ur = np.vstack([b[:437] for b in blocks])
lcs_matrix_ur = np.vstack(blocks)
lcs_matrix_ur.shape
```

```python
lcs_matrix_ur
```

```python
lcs_matrix = lcs_matrix_ur + (lcs_matrix_ur.T * (1-np.eye(lcs_matrix_ur.shape[0])))
lcs_matrix = lcs_matrix.astype("uint16")
```

```python
np.save(f"lcs_{dataset_name}.npy", lcs_matrix, allow_pickle=False)
```
